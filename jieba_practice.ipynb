```python
import os

flist = os.listdir('./res_gossip')
flist
```

```python
sb = ''

for f in flist:
    with open('./res_gossip/' + f, 'r', encoding='utf=8') as ff:
        sb += ff.read().split('---split---')[0].replace('\n', '')

sb
```

```python
sb_cut = jieba.cut(sb)
for i in sb_cut:
    print(i)
```

```python
sb_cut = jieba.cut(sb)
wc = {}
for w in sb_cut:
    if w in wc:
        wc[w] += 1
    else:
        wc[w] = 1
        
wc
```

```python
wc_list = []
for w in wc:
    wc_list.append((w, wc[w]))

wc_list
```

```python
[(w, wc[w]) for w in wc]
```

```python
wc_list.sort(key=lambda x: x[1], reverse=True)
wc_list
```

```python
stopword_list = []
with open('./stopword.txt', 'r', encoding='utf-8') as f:
    for w in f.readlines():
        stopword_list.append(w.replace('\n', ''))
stopword_list
```

```python
sb_cut = jieba.cut(sb)
wc = {}
for w in sb_cut:
    if len(w) < 2:
        continue
    if w in stopword_list:
        continue
    if w in wc:
        wc[w] += 1
    else:
        wc[w] = 1
        
wc_list = [(w, wc[w]) for w in wc]
wc_list.sort(key=lambda x: x[1], reverse=True)
wc_list
```
