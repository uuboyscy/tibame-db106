```python
import pandas as pd
import os
```

```python
flist = os.listdir('./res_gossip')
flist
```

```python
data_list = []
for f in flist:
    try:
        with open('./res_gossip' + '/' + f, encoding='utf-8') as s:
            tmp_str = s.read().split('---split---')[1]
            data_list.append(tmp_str)
    except:
        pass
for i in data_list:
    print(i)
```

```python
columns = ['文章編號', '推數', '噓數', '分數', '作者', '標題', '發表時間']
data = []

for i, d in enumerate(data_list):
    tmp_list = [str(i)] + d.split('\n')[1:]
    data.append(tmp_list)

data
```

```python
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
# Filter
def column_filter(s):
    return s.split(': ')[1]
```

```python
df2 = df.copy()
```

```python
df2['推數'].apply(column_filter)
```

```python
df2['推數'] = df2['推數'].apply(column_filter)
```

```python
df2
```

```python
for c in df:
    try:
        df[c] = df[c].apply(column_filter)
    except IndexError:
        pass
```
